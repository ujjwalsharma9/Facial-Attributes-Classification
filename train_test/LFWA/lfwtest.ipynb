```
from scipy import misc
import cv2
from keras.utils import np_utils
import numpy as np

def prepareTrain(feature):
  true_files=[]
  false_files=[]
  folder_path1='lfw/'
  for i in range(10000):
    folder_path2=df['person'][i].replace(" ","_")
    n=df['imagenum'][i]
    if n<10:
      file="000"+str(n)+".jpg"
    elif n<100:
      file="00"+str(n)+".jpg"
    else:
      file="0"+str(n)+".jpg"
  
    if df[feature][i] >=0:
      true_files.append(folder_path1+folder_path2+"/"+folder_path2+"_"+file)
    else:
      false_files.append(folder_path1+folder_path2+"/"+folder_path2+"_"+file)
  
  size=(len(true_files)+500-(len(true_files)%500))*2
  
  if size>10000:
    size=10000
  print(size)  
  file_paths=[]
  labels=[]
  a=0
  b=0
  c=0
  
  while c<size:
    if a<len(true_files):
      file_paths.append(true_files[a])
      labels.append(1)
      a=a+1
      c=c+1
      
    if b<len(false_files):
      file_paths.append(false_files[b])
      labels.append(0)
      b=b+1
      c=c+1
    
  
  images = [misc.imread(path) for path in file_paths]
  for i in range(len(images)):
    images[i]=cv2.resize(images[i],(128,128))
  
  images=np.asarray(images)
  images=images/255
 
  labels=np.asarray(labels)
  labels=np_utils.to_categorical(labels,2)
  
  return images,labels
  
  
  
  
  
  
```

```
sizes=[]
for i in range(25):
  
  X_train,y_train=prepareTrain(col_list[i])
  sizes.append(X_train.shape[0])
  print(X_train.shape[0])
  
  
```

```
from keras.models import Sequential
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Flatten
from keras.layers.core import Dropout
from keras.layers.core import Dense


model = Sequential()
model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu', 
                        input_shape=(128, 128, 3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=128, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))

model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(600, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.summary()

model.compile(loss='binary_crossentropy',
              optimizer='adam',metrics=['accuracy'])
```

```
from keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping
sizes=[]
for i in range(10,25):
  feature=col_list[i]
  print(i)
  print("Attribute: ",feature)
  X_train,y_train=prepareTrain(feature)
  sizes.append(X_train.shape[0])
  checkpointer = ModelCheckpoint(filepath='model.weights.best.hdf5', verbose=1, 
                               save_best_only=True)
  stop=EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=10,
                              verbose=1, mode='auto')
  hist = model.fit(X_train, y_train, batch_size=200, epochs=50,
          validation_split=0.2, callbacks=[checkpointer,stop], 
          verbose=2, shuffle=True) 
  model.load_weights('model.weights.best.hdf5')
  model.save('model_'+feature.replace(" ","_")+'.h5')
  print("Model Saved")
```

```
from scipy import misc
import cv2
from keras.utils import np_utils
import numpy as np

def prepareTest(feature,size):
  true_files=[]
  false_files=[]
  folder_path1='lfw/'
  for i in range(8000,13143):
    folder_path2=df['person'][i].replace(" ","_")
    n=df['imagenum'][i]
    if n<10:
      file="000"+str(n)+".jpg"
    elif n<100:
      file="00"+str(n)+".jpg"
    else:
      file="0"+str(n)+".jpg"
  
    if df[feature][i] >=0:
      true_files.append(folder_path1+folder_path2+"/"+folder_path2+"_"+file)
    else:
      false_files.append(folder_path1+folder_path2+"/"+folder_path2+"_"+file)
      
  a=0
  b=0
  c=0
  file_paths=[]
  labels=[]
  while c<size:
    if a<len(true_files):
      file_paths.append(true_files[a])
      labels.append(1)
      a=a+1
      c=c+1
        
    if b<len(false_files):
      file_paths.append(false_files[b])
      labels.append(0)
      b=b+1
      c=c+1
        
  images = [misc.imread(path) for path in file_paths]
  for i in range(len(images)):
    images[i]=cv2.resize(images[i],(128,128))
  
  images=np.asarray(images)
  images=images/255
 
  labels=np.asarray(labels)
  labels=np_utils.to_categorical(labels,2)
  
  return images,labels
  
  
```

```
from keras.models import load_model
s=0
t=0
acc=[]
for i in range(25):
  print("Attribute:",i)
  
  feature=col_list[i]
  print(feature)
  size=sizes[i]//2
  X_test,y_test=prepareTest(feature,size)
  model = load_model('model_'+feature.replace(" ","_")+'.h5')
  print(y_test[0])
  preds=model.predict(X_test)
  print(preds[0])
  preds=np.round(preds, 2)
  preds[preds>=0.5] = 1
  preds[preds<0.5] = 0
  score=np.sum(preds==y_test)/(preds==y_test).size
  print("Acc:", score*100) 
  acc.append(score*100)
  s=s+np.sum(preds==y_test)
  t=t+(preds==y_test).size
  
  
```

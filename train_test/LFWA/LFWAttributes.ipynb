```
url = 'http://www.cs.columbia.edu/CAVE/databases/pubfig/download/lfw_attributes.txt'  
wget.download(url, 'attributes.txt')  
```

```
text_file = open("attributes.txt", "r")
lines = text_file.readlines()
print(len(lines))
text_file.close()
```

```
len(lines[1])
```

```
lines[13144]
```

```
columns=lines[1].split("\t")
columns=columns[1:]
columns
```

```
import pandas as pd

df=pd.DataFrame(columns=columns,index=range(1,13144))
df
```

```
df.index
```

```
for i in range(1,13144):
  rows=lines[i+1].split("\t")
  rows[0].replace(" ", "_")
  rows[-1]=rows[-1].split('\n')[0]

  for j in range(1,len(rows)):
    rows[j]=float(rows[j])
    
  df.loc[i] =rows
  print(i)
```

```
df
```

```
df['person']
```

```
df.to_csv('LFW_Attributes.csv',float_format='%.4f%%')
```
